```python
# kofe mashina
database={
# gramm da berilgan omboridagi hom ashyo
"bunker":{
    "coffee":1000,
    "sugar":1000,
    "tea_ext":1000,
    "milk":1000,
    "chocolate":2000,
    "middle_cup":1000,
    "big_cup":500
},
# kassadagi pul
"bank":{
    "100000":5,
    "50000":10,
    "20000":20,
    "10000":20,
    "10000":20,
    "5000":200,
    "2000":500,
    "1000":500 },
# maxsulotlar narhlari
"price":{
    "Tea":3000,
    "Americano":7000,
    "Expresso":7000,
    "Cappuchino":8000,
    "Latte":8000,
    "Coffee with milk":8000,
    "Mokachino":10000,   
    "Chocolate":15000
},
# retsepti, {bunkerdagi hom ashyo nomi: kerakli miqdori gr}
"recipe":{'Tea': {'tea_ext': 5, 'middle_cup': 1},
         'Americano': {'coffee': 7, 'big_cup': 1},
         'Expresso': {'coffee': 7, 'middle_cup': 1},
         'Cappuchino': {'coffee': 7, 'milk': 7, 'big_cup': 1},
         'Latte': {'coffee': 7, 'milk': 21, 'middle_cup': 1},
         'Coffee with milk': {'coffee': 7, 'milk': 7, 'middle_cup': 1},
         'Mokachino': {'coffee': 7, 'milk': 14, 'chocolate': 7, 'big_cup': 1},
         'Chocolate': {'milk': 15, 'chocolate': 100, 'big_cup': 1}}
}
# bunkerda xom ashyo yetarlimi aniqlaydi
def is_inBunker(orders,bunker):
    for order in orders:
        for raw in database["recipe"][order]:
            
            if bunker[raw]-database["recipe"][order][raw]*orders[order]<0:
                return False
            bunker[raw]-=database["recipe"][order][raw]*orders[order]
    return bunker
            
# puli yetadimi tekshiradi oshsa qolganini qaytaradi
def able_toPay(orders):
    
    for order in orders:
        if money-database["price"][order]*orders[order]<0:
            
            return False
        else:
            
            return money-database["price"][order]*orders[order]
# qaytim bor yo`qligini aniqlaydi
def is_excess(orders):
    banknotes={}
    exc=able_toPay(orders)
    for i in database["bank"]:
        if 1<=exc//(int(i)) and database["bank"][i]>0:
            banknotes[i]=exc//(int(i))
            exc=exc%(int(i))
            if exc==0:
                break
    if exc!=0:
        return False
    else:
        return banknotes
# pul kiritishga ishlaydi
def take_m(m=0):
    global money
    if money==0:
        print("\n Ichimlik uchun",end=" ")
    money=int(input("pulni kiriting="))
    if money<0:
        return False
    money+=m
    show_menu(money)
# menyuni ekranga chiqaradi
def show(c):
    global database, money
    to_next="y"
    while to_next=="y":
        orders={}
        bunker_copy=database["bunker"].copy()
        for i in range(len(c)):
            print(f"{i+1}:{c[i]}={database['price'][c[i]]} so`m")
        order=int(input("Iltimos maxsulot nomerini kiriting="))
        while order>len(c):
            order=int(input("Iltimos maxsulot nomerini kiriting="))
#         for raw in recipe(c[order-1]):
#         bunker_copy[raw]=bunker_copy[raw]-recipe[c[order-1]][raw]
        items=int(input("Sonini kiriting="))
        while type(items)!=int or items<=0:
            items=int(input("Sonini kiriting="))
        if c[order-1]!="Chocolate" and bunker_copy["sugar"]>=items*5:
            y_n=input("Shakar bilanmi? y/n")
            while y_n!="y" and y_n!="n":
                y_n=input("Shakar bilanmi? y/n")
            if y_n=="y":
                bunker_copy["sugar"]-=items*5
        orders[c[order-1]]=items 
        qaytim=able_toPay(orders)
        if qaytim or qaytim==0:
            if type(is_excess(orders))==dict:
                if is_inBunker(orders,bunker_copy):
                    print(f"marxamat {orders} ")
                    money=able_toPay(orders)
                    database["bunker"]=bunker_copy.copy()
                    orders={}
                    to_next=input("yana boshqa buyurtma berasizmi? y/n") 
                    while to_next!="y" and to_next!="n":
                        to_next=input("yana boshqa buyurtma berasizmi? y/n") 
                    if to_next=="y":
                        show_menu(money)
                    elif money!=0:                    
                        print("qaytimingiz",money)
                        money=0
                        take_m()
                    elif money==0:
                        take_m()
                else:
                    print("Afsuski sizni buyurtmangiz qolmagan yoki siz so`ragandan ozroq bor, iltimos boshqa kiriting")
                    show(c)
            else:
                print("Qaytishga pul kamlik qiladi,iltimos boshqa pul kiriting")
                print(f"Mana pulingiz {money} so`m")
                money=0
                take_m()
        else:
            t=input(f"Pulingiz {money}so`m to`lovga yetmaydi boshqa buyurtma berasizmi? y/n")
            while t!="y" and t!="n":
                t=input(f"Pulingiz {money}so`m to`lovga yetmaydi boshqa buyurtma berasizmi? y/n")
            if t=="y":
                show_menu(money)
            else:
                take_m()
# puliga yarasha maxsulotlarni ko`rsatadi   
def show_menu(a):
    global money
    c=[]
    if a<database["price"]["Tea"]:
        y_n=input(f"Pulingiz {a}so`m to`lovga yetmaydi, yana pul kiritasizmi? y/n")
        while y_n!="y" and y_n!="n":
            y_n=input("Pulingiz yetmaydi, yana pul kiritasizmi? y/n")
        if y_n=="y":
            take_m(a)
        else:
            money=0
            print("marxamat qaytimingiz=",a)
            take_m()
    bunker=database["bunker"].copy()        
    if database["price"]["Tea"]<=a and is_inBunker({"Tea":1},bunker):
        c.append("Tea")
    if database["price"]["Americano"]<=a and is_inBunker({"Americano":1},bunker):
        c.append("Americano"); c.append("Expresso") 
    if database["price"]["Cappuchino"]<=a:
        if is_inBunker({"Cappuchino":1},bunker):
            c.append("Cappuchino")
        if is_inBunker({"Latte":1},bunker):
            c.append("Latte")
        if is_inBunker({"Coffee with milk":1},bunker):
            c.append("Coffee with milk")
    if database["price"]["Mokachino"]<=a and is_inBunker({"Mokachino":1},bunker):
        c.append("Mokachino")
    if database["price"]["Chocolate"]<=a and is_inBunker({"Chocolate":1},bunker):
        c.append("Chocolate")
    show(c)
money=0
take_m()
```
